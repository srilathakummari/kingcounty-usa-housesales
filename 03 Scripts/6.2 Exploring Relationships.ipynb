### Categorical Plots

#### Creating a categorical variable that splits the "Price" column into categories

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the price ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df_clean_hs['price'], bins = 10, kde = True)
```

```python
# Distribution of house sales price by densitiy 
plt.figure(figsize=(16,8))
sns.distplot(df_clean_hs['price'], hist=True, kde=True, 
             bins=int(180/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
```

```python
df_clean_hs.loc[df_clean_hs['price'] < 400000, 'Price category'] = 'Low price'
```

```python
df_clean_hs.loc[(df_clean_hs['price'] >= 600000) & (df_clean_hs['price'] < 800000), 'Price category'] = 'Middle price'
```

```python
df_clean_hs.loc[df_clean_hs['price'] >= 900000, 'Price category'] = 'High price'
```

```python
df_clean_hs['Price category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="grade", y="price", hue="Price category", data=df_clean_hs)
```
